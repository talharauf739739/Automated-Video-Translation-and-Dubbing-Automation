```python
import os
import requests
import whisper
from pydub import AudioSegment

def preprocess_audio(wav_path):
    """Normalize audio to ensure consistent volume."""
    audio = AudioSegment.from_wav(wav_path)
    normalized_audio = audio.normalize()
    processed_wav_path = wav_path.replace(".wav", "_processed.wav")
    normalized_audio.export(processed_wav_path, format="wav")
    return processed_wav_path

def split_audio(wav_path, chunk_length_ms=300000):
    """Split long audio into chunks of specified length (default: 5 minutes)."""
    audio = AudioSegment.from_wav(wav_path)
    chunks = []
    for i in range(0, len(audio), chunk_length_ms):
        start_time = i
        end_time = min(i + chunk_length_ms, len(audio))
        chunk = audio[start_time:end_time]
        chunks.append((start_time / 1000, end_time / 1000, chunk))  # Timestamps in seconds
    return chunks

def transcribe_with_whisper(audio_chunk, model):
    """Transcribe a single audio chunk with Whisper."""
    result = model.transcribe(audio_chunk, verbose=False)
    return result.get("segments", [])

def transcribe_with_deepgram(audio_chunk_path, deepgram_api_key):
    """Transcribe a single audio chunk with Deepgram."""
    with open(audio_chunk_path, 'rb') as f:
        response = requests.post(
            f"https://api.deepgram.com/v1/listen?punctuate=true&utterances=true",
            headers={"Authorization": f"Bearer {deepgram_api_key}"},
            files={"file": f}
        )
        if response.status_code == 200:
            return response.json().get("results", {}).get("utterances", [])
        else:
            raise RuntimeError(f"Deepgram API error: {response.status_code}")

def process_long_audio(wav_path, txt_dir, method="whisper", chunk_length_ms=300000, deepgram_api_key=None):
    """
    Process a long audio file for transcription using Whisper or Deepgram.

    Args:
        wav_path: Path to the input .wav file.
        txt_dir: Directory to save transcriptions.
        method: "whisper" or "deepgram".
        chunk_length_ms: Length of each audio chunk in milliseconds.
        deepgram_api_key: API key for Deepgram (required if method="deepgram").
    """
    # Ensure output directory exists
    if not os.path.exists(txt_dir):
        os.mkdir(txt_dir)

    # Preprocess and split audio
    processed_wav_path = preprocess_audio(wav_path)
    audio_chunks = split_audio(processed_wav_path, chunk_length_ms)

    # Load Whisper model if selected
    model = whisper.load_model("base") if method == "whisper" else None

    # Prepare output transcription file
    txt_path = os.path.join(txt_dir, os.path.basename(wav_path)[:-4] + "_timestamps.txt")
    with open(txt_path, "w") as text_file:
        for start, end, chunk in audio_chunks:
            chunk_path = f"temp_chunk_{start:.2f}_{end:.2f}.wav"
            chunk.export(chunk_path, format="wav")

            # Transcribe based on the selected method
            if method == "whisper":
                segments = transcribe_with_whisper(chunk_path, model)
                for segment in segments:
                    # Adjust timestamps relative to the original file
                    adjusted_start = start + segment["start"]
                    adjusted_end = start + segment["end"]
                    text = segment["text"]
                    text_file.write(f"[{adjusted_start:.2f} - {adjusted_end:.2f}] {text}\n")
            elif method == "deepgram":
                utterances = transcribe_with_deepgram(chunk_path, deepgram_api_key)
                for utterance in utterances:
                    # Adjust timestamps relative to the original file
                    adjusted_start = start + utterance["start"]
                    adjusted_end = start + utterance["end"]
                    text = utterance["transcript"]
                    text_file.write(f"[{adjusted_start:.2f} - {adjusted_end:.2f}] {text}\n")
            else:
                raise ValueError("Invalid method. Choose 'whisper' or 'deepgram'.")

            # Remove the temporary chunk file
            os.remove(chunk_path)

    print(f"Transcription saved to: {txt_path}")

if __name__ == "__main__":
    # Input directory and parameters
    wav_path = r"/content/drive/MyDrive/Dutch wav/Dutch_Hockey.wav"
    txt_dir = r"/content/"
    deepgram_api_key = "5b8b243e83906934f93c55d141d9d95315239b12"

    # Choose transcription method: Whisper or Deepgram
    method = "whisper"  # or "deepgram"

    # Process long audio
    process_long_audio(wav_path, txt_dir, method=method, chunk_length_ms=300000, deepgram_api_key=deepgram_api_key)
```

```python
from transformers import MarianMTModel, MarianTokenizer

def read_transcription(file_path):
    """Reads transcription with timestamps from the .txt file."""
    transcription = []
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            # Parse lines of format: [start - end] text
            parts = line.strip().split("]")
            if len(parts) == 2:
                time_range, text = parts
                start, end = map(float, time_range[1:].split(" - "))
                transcription.append({"start": start, "end": end, "text": text})
    return transcription

def translate_text(text, tokenizer, model):
    """Translate text using the given MarianMT model and tokenizer."""
    inputs = tokenizer.encode(text, return_tensors="pt", padding=True, truncation=True)
    translated = model.generate(inputs, max_length=250, num_beams=8, repetition_penalty=1.5, length_penalty=1.2, early_stopping=True)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

def translate_transcription(transcription, txt_dir, nl_to_en_model, nl_to_en_tokenizer, en_to_de_model, en_to_de_tokenizer):
    """Translate transcription and save translations to .txt files."""
    dutch_to_eng_path = f"{txt_dir}/dutch_to_eng.txt"
    dutch_to_german_path = f"{txt_dir}/dutch_to_german.txt"

    with open(dutch_to_eng_path, "w", encoding="utf-8") as eng_file, open(dutch_to_german_path, "w", encoding="utf-8") as ger_file:
        for segment in transcription:
            start = segment["start"]
            end = segment["end"]
            dutch_text = segment["text"]

            # Dutch → English
            english_text = translate_text(dutch_text, nl_to_en_tokenizer, nl_to_en_model)
            eng_file.write(f"[{start:.2f} - {end:.2f}] {english_text}\n")

            # English → German
            german_text = translate_text(english_text, en_to_de_tokenizer, en_to_de_model)
            ger_file.write(f"[{start:.2f} - {end:.2f}] {german_text}\n")

    print(f"Translations saved to:\n{dutch_to_eng_path}\n{dutch_to_german_path}")

if __name__ == "__main__":
    # Model loading
    nl_to_en_model_name = "Helsinki-NLP/opus-mt-nl-en"
    en_to_de_model_name = "Helsinki-NLP/opus-mt-en-de"

    nl_to_en_tokenizer = MarianTokenizer.from_pretrained(nl_to_en_model_name)
    nl_to_en_model = MarianMTModel.from_pretrained(nl_to_en_model_name)

    en_to_de_tokenizer = MarianTokenizer.from_pretrained(en_to_de_model_name)
    en_to_de_model = MarianMTModel.from_pretrained(en_to_de_model_name)

    # Transcription file
    transcription_path = "/content/Dutch_Hockey_timestamps.txt"

    # Directory for translated outputs
    output_dir = "/content"

    # Read transcription and perform translation
    transcription = read_transcription(transcription_path)
    translate_transcription(transcription, output_dir, nl_to_en_model, nl_to_en_tokenizer, en_to_de_model, en_to_de_tokenizer)
```

```python
import os
from models import build_model
import torch
from pydub import AudioSegment
from kokoro import generate

# Load the Kokoro-82M model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
MODEL = build_model('/content/Kokoro-82M/kokoro-v0_19.pth', device)
VOICE_NAME = [
    'af', # Default voice is a 50-50 mix of Bella & Sarah
    'af_bella', 'af_sarah', 'am_adam', 'am_michael',
    'bf_emma', 'bf_isabella', 'bm_george', 'bm_lewis',
    'af_nicole', 'af_sky',
][3]
VOICEPACK = torch.load(f'voices/{VOICE_NAME}.pt', weights_only=True).to(device)

def parse_translation_file(file_path):
    """Parse the timestamped translation file."""
    segments = []
    with open(file_path, "r") as f:
        for line in f:
            if line.strip():
                timestamp, text = line.strip().split("]", 1)
                start, end = map(float, timestamp.strip("[").split(" - "))
                segments.append((start, end, text.strip()))
    return segments

def adjust_speed(audio, target_duration):
    """Adjust playback speed to match the target duration."""
    current_duration = len(audio) / 1000.0
    if current_duration == target_duration:
        return audio
    speed_factor = current_duration / target_duration
    return audio.speedup(playback_speed=speed_factor)

def generate_tts_for_segment(model, voicepack, lang, text, duration):
    """Generate TTS audio for a segment."""
    audio, _ = generate(model, text, voicepack, lang=lang)
    audio_segment = AudioSegment(data=audio.tobytes(), frame_rate=24000, sample_width=2, channels=1)
    return adjust_speed(audio_segment, duration)

def synthesize_with_timestamps(translations_path, output_audio_path):
    """Generate and save TTS audio aligned with original timestamps."""
    segments = parse_translation_file(translations_path)
    final_audio = AudioSegment.silent(duration=0)

    for start, end, text in segments:
        duration = end - start
        tts_audio = generate_tts_for_segment(MODEL, VOICEPACK, lang=VOICE_NAME[0], text=text, duration=duration)
        final_audio += tts_audio

    final_audio.export(output_audio_path, format="wav")
    print(f"Synthesized audio saved to {output_audio_path}")

# Paths
translations_path = "/content/dutch_to_eng.txt"
output_audio_path = "/content/dutch_to_eng_tts_aligned.wav"

# Run synthesis
synthesize_with_timestamps(translations_path, output_audio_path)
```
